```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
# Sample sizes
# Sample sizes
sample_sizes1 = [500, 200, 150, 100]
sample_sizes2 = [1000, 500, 200, 150, 100] 

# Error values for each sample size
errors1 = {
    500: [0.8268, 0.8099, 0.8168, 0.8418, 0.8469],
    200: [1.1239, 1.1099, 1.0999, 1.1020, 1.0987],
    150: [1.1220, 1.1364, 1.1328, 1.1318, 1.1337],
    100: [1.2271, 1.2116, 1.2263, 1.2143, 1.2141]
}

errors2 = {
    1000: [0.8768, 0.8099, 0.8168, 0.8418, 0.8869],
    500: [1.4819, 1.4289, 1.4820, 1.4901, 1.4627],
    200: [3.3040, 3.2232, 3.6117, 3.4180, 3.3049],
    150: [4.3201, 4.5999, 4.4860, 4.5906, 4.4078],
    100: [5.2045, 5.3510, 5.2769, 5.2085, 5.4064]
}


# Calculate mean and std for each sample size
mean_errors1 = [np.mean(errors1[size]) for size in sample_sizes1]
std_errors1 = [np.std(errors1[size]) for size in sample_sizes1]

mean_errors2 = [np.mean(errors2[size]) for size in sample_sizes2]
std_errors2 = [np.std(errors2[size]) for size in sample_sizes2]

# Set font properties
font = {'family': 'serif', 'size': 16}
plt.rc('font', **font)
plt.rc('lines', linewidth=1.5)  # Increase line width for better visibility

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Uniformly spaced x-ticks
xticks1 = np.linspace(min(sample_sizes1), max(sample_sizes1), 5)
xticks2 = np.linspace(min(sample_sizes2), max(sample_sizes2), 5)

# Plotting first subplot
axes[0].errorbar(sample_sizes1, mean_errors1, ecolor='black', yerr=std_errors1, fmt='-o', capsize=5, linewidth=2.5, linestyle='--', color='green')
axes[0].set_xlabel('Training Input Size ($S_\mathrm{exact\_inference}$)', fontsize=16)
axes[0].set_ylabel('Mean $L^2$ Error (%)', fontsize=16)
axes[0].set_ylim(min(mean_errors1) - 0.1, max(mean_errors1) + 0.1)  # Adjust y-axis limits
axes[0].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
axes[0].set_yticks(np.round(np.linspace(min(mean_errors1) - 0.1, max(mean_errors1) + 0.1, 6), 4))  # Show y-ticks with decimal precision
axes[0].set_xticks(xticks1)  # Uniform x-ticks
# axes[0].grid(True, which='both', linestyle='--', linewidth=0.5)
axes[0].text(0.5, -0.20, '(a)', transform=axes[0].transAxes, ha='center', va='center', fontsize=18)

# Plotting second subplot
axes[1].errorbar(sample_sizes2, mean_errors2, yerr=std_errors2, fmt='-o', capsize=5, linewidth=2.5, linestyle='--', color='purple',ecolor='black')
axes[1].set_xlabel('Training Input Size ($S_\mathrm{SDD}$)', fontsize=16)
axes[1].set_ylim(min(mean_errors2) - 0.1, max(mean_errors2) + max(std_errors2) + 0.1)  # Adjust y-axis limits
axes[1].yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
axes[1].set_yticks(np.round(np.linspace(min(mean_errors2) - 0.1, max(mean_errors2) + max(std_errors2) + 0.1, 6), 4))  # Show y-ticks with decimal precision
axes[1].set_xticks(xticks2)  # Uniform x-ticks
# axes[1].grid(True, which='both', linestyle='--', linewidth=0.5)
axes[1].text(0.5, -0.20, '(b)', transform=axes[1].transAxes, ha='center', va='center', fontsize=18)

# Adjust layout and show the plot
plt.tight_layout()
plt.savefig('/home/user/Documents/WaKeGP/experiments/C2_wa/results/wa_error.pdf', format='pdf', dpi=600, bbox_inches='tight')

plt.show()
```
